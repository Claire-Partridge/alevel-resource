```python
from IPython.display import display, Math, Latex, HTML, Image
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

<font color = purple> $\mathtt{Import\ your\ Pressure\ Data\ from\ PhyPhox\ as\ a\ Pandas\ Dataframe}$ </font>

```python
#importing the phyphox data
#Match the file name of your exported data to the name
#in the quotation marks

dataframe = pd.read_csv("pressure.csv", header=0)
#                        ^^ file name.csv


xData = np.array(dataframe.time)
yData = np.array(dataframe.pressure)
```

```python

plt.plot(xData,yData, label = 'phyphox data', color = 'pink', marker = 'o', mfc = 'm')
plt.xlabel('\ntime / s')
plt.ylabel('pressure / hPa\n')
plt.grid(color = 'purple', alpha = 0.25, linestyle = '--')
plt.legend()
plt.title('Pressure.csv import\n')
plt.show();
```
